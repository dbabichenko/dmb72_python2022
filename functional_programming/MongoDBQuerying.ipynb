### Connecting to MongoDB 

```python
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["sakila"]
mycol = mydb["Sakila_actors"]

myquery = { "FirstName": "NICK" }

mydoc = mycol.find(myquery)

for x in mydoc:
    print(x)
```

### MongoDB Find

```python
# Find documents where the name starts with the letter "N" or higher:
myquery = { "LastName": { "$gt": "N" } }

mydoc = mycol.find(myquery)

for x in mydoc:
    print(x)
```

```python
# Filter With Regular Expressions: 
# Find documents where the name starts with the letter "N":
myquery = { "LastName": { "$regex": "^N" } }

mydoc = mycol.find(myquery)

for x in mydoc:
    print(x)
```

```python
# Find the first document in the customers collection
x = mycol.find_one()

print(x)
```

**Find All**
* To select data from a table in MongoDB, we can also use the find() method.
* The find() method returns all occurrences in the selection.
* The first parameter of the find() method is a query object. 
* In this example we use an empty query object, which selects all documents in the collection.
* No parameters in the find() method gives you the same result as SELECT in MySQL.

```python
for x in mycol.find():
    print(x)
```

**Return Only Some Fields**

* The second parameter of the find() method is an object describing which fields to include in the result.
* This parameter is optional, and if omitted, all fields will be included in the result.



```python
# Return only the names, not the _ids:
for x in mycol.find({},{ "_id": 0, "FirstName": 1, "LastName": 1 }):
    print(x)
```

```python
# Exclude "LastName" from the result:

for x in mycol.find({},{ "address": 0 }):
    print(x)
```

MongoDB API uses JavaScript to create map and reduce functions.  The example below demonstrates a JavaScript function that gets evaluated by MongoDB engine and returns _name_ and _marks_ from the data that it is applied to.

<div style="border: 1px solid #000000; padding: 10px; font-style: italic">
var map = function() {emit(this.name,this.marks);};
</div>

In order for us to wrap a JavaScript function in a Python call, we can use the Code class to evaluate a JavaScript string:

<div style="border: 1px solid #000000; padding: 10px; font-style: italic">
map = Code("var map = function() {emit(this.name,this.marks);};")

<div style="border: 1px solid #000000; padding: 10px; font-style: italic">

And, just like with the map example, we wrap the JavaScript reduce() function with Python's Code()

result = mydb.mycol.map_reduce(map,reduce,"out")

### Aggregation Pipeline

An aggregation pipeline consists of one or more stages that process documents:
* Each stage performs an operation on the input documents. For example, a stage can filter documents, group documents, and calculate values.
* The documents that are output from a stage are passed to the next stage.
* An aggregation pipeline can return results for groups of documents. For example, return the total, average, maximum, and minimum values.

```python
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["pizza"]
col = db["orders"]
```

**Calculate Total Order Quantity**

The following aggregation pipeline example contains two stages and returns the total order quantity of medium size pizzas grouped by pizza name:


```python
# Stage 1: Filter pizza order documents by pizza size
# Stage 2: Group remaining documents by pizza name and calculate total quantity
pipeline = [
   {
      "$match": { "size": "medium" }
   },
   {
      "$group": { "_id": "$name", "totalQuantity": { "$sum": "$quantity" } }
   }
]

result = col.aggregate(pipeline)
```

SELECT name, SUM(totalQuantity) FROM orders
GROUP BY name 
WHERE size = 'medium'
