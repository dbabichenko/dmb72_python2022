```python
import csv
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = open('9_11.csv', 'r')
data = csv.reader(f)
```

```python
G = nx.Graph()

for line in data:
    #print(line)
    G.add_edge(line[0], line[1], weight=line[2], conf=line[3])
```

```python
G.nodes
```

```python
deg_cent = nx.degree_centrality(G)

{k: v for k, v in sorted(deg_cent.items(), key=lambda item: item[1])}
```

```python
edges = G.edges()
weights = [G[u][v]['weight'] for u,v in edges]
conf = [G[u][v]['conf'] for u,v in edges]

print(weights)
```

```python
plt.figure(figsize=(20, 14), dpi=80)
nx.draw_networkx(G, node_size=500)
```
