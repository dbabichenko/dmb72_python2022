```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/viz.csv')
df.head()
```

```python
G = nx.from_pandas_edgelist(df, source="Source", target="Target", edge_attr="Weight")
```

```python
plt.figure(figsize=(20, 10))
nx.draw(G)
```

```python
plt.figure(figsize=(20, 10))
nx.draw_networkx(G)
```

```python
plt.figure(figsize=(20, 10))
nx.draw_circular(G)
```

```python
plt.figure(figsize=(20, 10))
nx.draw_random(G)
```

```python
nx.draw_spring(G)
```

```python
#import pyvis
from pyvis.network import Network
```

```python
net = Network(notebook=True)
net.from_nx(G)
net.show("test.html")
```
